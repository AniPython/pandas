```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3])
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4
```

```python
plt.figure(facecolor='white')
plt.plot(x, y1, label='1')
plt.plot(x, y2, label='2')
plt.plot(x, y3, label='3')
plt.plot(x, y4, label='4')
plt.legend()
```

```python
fig, ax = plt.subplots(
    facecolor='white'
)
ax.plot(x, y1, label='1')
ax.plot(x, y2, label='2')
ax.plot(x, y3, label='3')
ax.plot(x, y4, label='4')
ax.legend()
```

```python
plt.figure(facecolor='white')

plt.subplot(221)
plt.plot(x, y1, label='1')
plt.xticks(fontsize=20)

plt.subplot(222)
plt.plot(x, y2, label='2', color='r')
plt.legend()

plt.subplot(223)
plt.plot(x, y3, label='3')

plt.subplot(224)
plt.plot(x, y4, label='4')
plt.legend()

plt.tight_layout()
```

```python
_, axs = plt.subplots(
    2, 2, facecolor='white',
    figsize=(10, 8)
)
axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[1, 0].plot(x, y3)
axs[1, 1].plot(x, y4)
```

```python
_, axs = plt.subplots(
    4, 1, facecolor='white',
    figsize=(3, 8),
    sharex='all'
)
axs[0].plot(x, y1)
axs[1].plot(x, y2)
axs[2].plot(x, y3)
axs[3].plot(x, y4)
```
