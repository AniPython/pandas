#### figure-level 和 axes-level 的区别
relplot(), catplot() 等函数: figure-level
返回: seaborn.FacetGrid
lineplot(), scatterplot() 等函数: axes-level
返回: matplotlib.Axes

#### FacetGrid 方法
- facet_axis(row_i, col_j): 获取指定轴
- set(
    xlim,
    ylim,
    xlabel,
    xticklabels,
    ylabel,
    yticklabels
    title):           在每个子图轴上设置属性
- set_axis_labels():  设置轴标签
- set_titles():       绘制标题
- set_xlabels():      x轴标签
- set_xticklabels():  x轴刻度标签
- set_ylabels():      y轴标签
- set_yticklabels():  y轴刻度标签
#### FacetGrid 属性
- ax:   单轴
- axes: 多轴(axes.flat)

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
g = sns.relplot(
    kind='line',
    x='day',
    y='total_bill',
    col='sex',
    row='time',
    data=tips,
    # ci=None  # ci 参数弃用, 使用 `errorbar=None`
    errorbar=None
)
# for i, ax in enumerate(g.axes.flat):
#     ax.set_title(str(i))
# ax = g.facet_axis(row_i=1, col_j=0)
# ax.set_title('AAA')
g.set(ylim=[10, 30])
g.set_titles('AAA')
```

```python
ax = sns.lineplot(
    x='day',
    y='total_bill',
    data=tips,
    # ci=None # ci 参数弃用, 使用 `errorbar=None`
    errorbar=None
)
# ax.set_xlabel('xxx')
# ax.set_title('BBB')
sns.scatterplot(
    x='day',
    y='tip',
    data=tips,
    # ci=None, # ci 参数弃用
    ax=ax
)
```
