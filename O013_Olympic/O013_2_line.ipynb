#### line 线图: 强调连续性
- seaborn.relplot(kind='line'): figure-level
- seaborn.lineplot(): axes-level

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('./input/athlete_events.csv')
df2 = pd.read_csv('./input/noc_regions.csv')
df = pd.merge(
    df1, df2,
    on='NOC',
    how='left',
    validate='m:1')
df.info()
```

```python
sns.set(
    context='talk',  # paper, notebook, talk, poster
    style='whitegrid',  # darkgrid, whitegrid, dark, white, ticks
    palette='muted',  # deep,muted,bright,pastel,dark,colorblind,hls,husl,cm
    rc={
        'figure.figsize': [8, 6]
    }
)
```

```python
sns.lineplot(
    x='Year',
    y='Height',
    data=df
)
```

```python
sns.lineplot(
    x='Year',
    y='Height',
    data=df,
    hue='Season'
)
```

```python
sns.lineplot(
    x='Year',
    y='Height',
    data=df,
    hue='Season',
    style='Sex'
)
```

```python
sns.relplot(
    kind='line',
    x='Year',
    y='Height',
    data=df,
    hue='Season',
    style='Sex',
    aspect=1.5
)
```
