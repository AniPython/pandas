```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(
    context='talk',  # paper, notebook, talk, poster
    style='darkgrid',  # darkgrid, whitegrid, dark, white, ticks
    palette='muted',  # deep,muted,bright,pastel,dark,colorblind,hls,husl,cm
    rc={
        'figure.figsize': [8, 6]
    }
)
```

```python
df1 = pd.read_csv('./input/athlete_events.csv')
df2 = pd.read_csv('./input/noc_regions.csv')
df = pd.merge(
    df1, df2,
    on='NOC',
    how='left',
    validate='m:1')
```

```python
df = df.query('Year>=1984')
df.drop_duplicates(['ID', 'Games'])
df.info()
```

```python
sns.distplot(
    a=df['Weight'],
    bins=None,
    hist=True,
    kde=True, 
    rug=True)
```

```python
plt.figure(figsize=(18, 6))
ax = sns.distplot(
    a=df['Weight'],
    bins=None,
    hist=True,
    kde=False)
ax.xaxis.set_major_locator(
    plt.MultipleLocator(5)
)
ax.tick_params(rotation=90,axis='x')
```

```python
plt.figure(figsize=(18, 6))
ax = sns.distplot(
    a=df['Weight'],
    bins=20,
    hist=True,
    kde=False)
ax.xaxis.set_major_locator(
    plt.MultipleLocator(5)
)
ax.tick_params(rotation=90,axis='x')
```

```python
plt.figure(figsize=(18, 6))
ax = sns.distplot(
    a=df['Weight'],
    bins='doane',
    hist=True,
    kde=False)
ax.xaxis.set_major_locator(
    plt.MultipleLocator(5)
)
ax.tick_params(rotation=90,axis='x')
```

```python
plt.figure(figsize=(18, 6))
ax = sns.distplot(
    a=df['Weight'],
    bins='sturges',
    hist=True,
    kde=False)
ax.xaxis.set_major_locator(
    plt.MultipleLocator(5)
)
ax.tick_params(rotation=90,axis='x')
```

```python
plt.figure(figsize=(18, 6))
ax = sns.distplot(
    a=df['Weight'],
    bins='stone',
    hist=True,
    kde=False)
ax.xaxis.set_major_locator(
    plt.MultipleLocator(5)
)
ax.tick_params(rotation=90,axis='x')
```
