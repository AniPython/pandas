#### 分布图: 密度图/核密度估计
- seaborn.kdeplot(): axes-level

```python
import pandas as pd
import seaborn as sns
```

```python
sns.set(
    context='talk',  # paper, notebook, talk, poster
    style='darkgrid',  # darkgrid, whitegrid, dark, white, ticks
    palette='muted',  # deep,muted,bright,pastel,dark,colorblind,hls,husl,cm
    rc={
        'figure.figsize': [8, 6]
    }
)
```

```python
df1 = pd.read_csv('./input/athlete_events.csv')
df2 = pd.read_csv('./input/noc_regions.csv')
df = pd.merge(
    df1, df2,
    on='NOC',
    how='left',
    validate='m:1')
```

```python
df = df.query('Year>=1984')
df = df.drop_duplicates(['ID', 'Games'])
df.info()
```

```python
sns.kdeplot(
    data=df['Age']
)
```

```python
sns.kdeplot(
    data=df['Age'],
    # shade=True  # 警告, 将弃用, fill=True
    fill=True
)
```

```python
sns.kdeplot(
    data=df.query('Sex=="M"')['Age'],
    shade=True,
    color='b',
    label="M"
)
sns.kdeplot(
    data=df.query('Sex=="F"')['Age'],
    shade=True,
    color='r',
    label="F"
)
```

```python
sns.kdeplot(
    data=df.query('Sex=="F"')['Age'],
    shade=True,
    color='r',
    label="F",
    vertical=True
)
```
