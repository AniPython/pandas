```python
# https://matplotlib.org/tutorials/introductory/customizing.html#matplotlibrc-sample
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif']=['Hiragino Sans GB']  # win: ['SimHei']
plt.rcParams['figure.facecolor'] = 'y'
plt.rcParams['axes.facecolor'] = (0.8, 0.9, 0.8)
plt.rcParams['axes.titlesize'] = 36
plt.rcParams['axes.labelsize'] = 30
plt.rcParams['xtick.labelsize'] = 24
plt.rcParams['ytick.labelsize'] = 'large'
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['lines.linestyle'] = '--'
```

```python
x = np.array([1, 2, 3])
y2 = x ** 2
y4 = x ** 4
```

```python
plt.plot(x, y2, label='2')
plt.plot(x, y4, label='4')

plt.title('我是标题')
plt.xlabel('X轴')
plt.ylabel('Y轴')

plt.legend()
```

```python
_, axs = plt.subplots(
    2, 1,
    figsize=(6, 8))
axs[0].plot(x, y2, label='2')
axs[0].set_title('我是平方')
axs[0].set_xlabel('X轴')
axs[0].set_ylabel('Y轴')

axs[1].plot(x, y4, label='4')
axs[1].set_title('我是四次方')
axs[1].set_xlabel('X轴')
axs[1].set_ylabel('Y轴')

plt.tight_layout()
```
