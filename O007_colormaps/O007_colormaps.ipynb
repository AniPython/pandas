```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(1, 30, size=(3,3))
data
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-white')
```

```python
df = pd.DataFrame(data)
```

```python
df.plot()
```

```python
df.plot(colormap=plt.cm.spring)
```

```python
df.plot(colormap='Set3')
```

```python
from matplotlib.colors import ListedColormap
my_cm = ListedColormap(['orange', 'yellow', 'blue'])
df.plot(colormap=my_cm)
```
