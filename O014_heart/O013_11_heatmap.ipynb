#### heatmap 热力图
- seaborn.heatmap(): axes-level

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(
    context='talk',  # paper, notebook, talk, poster
    style='darkgrid',  # darkgrid, whitegrid, dark, white, ticks
    palette='muted',  # deep,muted,bright,pastel,dark,colorblind,hls,husl,cm
    rc={
        'figure.figsize': [8, 6]
    }
)
```

```python
data = np.random.randint(1, 11, size=(5, 5))
data
```

```python
sns.heatmap(data)
```

```python
sns.heatmap(
    data,
    annot=True,
    cmap='Greens'
)
```

```python
sns.heatmap(
    data,
    annot=True,
    cmap='Greens_r',
    vmin=1,
    vmax=10,
    fmt='.1f'
)
```

```python
sns.heatmap(
    data,
    annot=True,
    cmap='coolwarm',
    vmin=1,
    vmax=10,
    fmt='.1f',
    center=5
)
```

```python
sns.heatmap(
    data,
    annot=True,
    cmap='coolwarm',
    vmin=1,
    vmax=10,
    fmt='.1f',
    center=8
)
```

```python
sns.heatmap(
    data,
    annot=True,
    cbar=False
)
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
corr = df.corr()
corr
```

```python
plt.figure(figsize=[16, 10])
sns.heatmap(
    corr,
    cmap='coolwarm',
    vmin=-1,
    vmax=1,
    annot=True,
    fmt='.2f',
    center=0
)
```
